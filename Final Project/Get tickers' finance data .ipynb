```python
import pandas as pd
import numpy as np
import datetime as dt
from pandas_datareader import data as pdr
from scipy.stats import norm, t
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
startDate = "2020-01-01"
endDate = "2022-11-17"

tsla = yf.Ticker("TSLA")
tslaData = tsla.history(start=startDate, end=endDate)
tslaData=tslaData.tail(100)
tslaData['Ticker'] = 'TSLA'

aapl = yf.Ticker("AAPL")
aaplData = aapl.history(start=startDate, end=endDate)
aaplData=aaplData.tail(100)
aaplData['Ticker'] = 'AAPL'

amzn = yf.Ticker("AMZN")
amznData=amzn.history(start=startDate, end=endDate)
amznData=amznData.tail(100)
amznData['Ticker']="AMZN"

msft = yf.Ticker("MSFT")
msftData=msft.history(start=startDate, end=endDate)
msftData=msftData.tail(100)
msftData['Ticker']="MSFT"

spy = yf.Ticker("SPY")
spyData=spy.history(start=startDate, end=endDate)
spyData=spyData.tail(100)
spyData['Ticker']="SPY"

ko = yf.Ticker("KO")
koData=ko.history(start=startDate, end=endDate)
koData=koData.tail(100)
koData['Ticker']="KO"

t = yf.Ticker("T")
tData=t.history(start=startDate, end=endDate)
tData=tData.tail(100)
tData['Ticker']="T"

intc = yf.Ticker("INTC")
intcData=intc.history(start=startDate, end=endDate)
intcData=intcData.tail(100)
intcData['Ticker']="INTC"

tgt = yf.Ticker("TGT")
tgtData=tgt.history(start=startDate, end=endDate)
tgtData=tgtData.tail(100)
tgtData['Ticker']="TGT"

goog = yf.Ticker("GOOG")
googData=goog.history(start=startDate, end=endDate)
googData=googData.tail(100)
googData['Ticker']="GOOG"

allData = pd.concat([tslaData,aaplData,amznData,msftData,spyData,koData, tData, intcData,tgtData,googData], axis=0)
```

```python
allData
```

```python
allData=allData.reset_index()
allData
```

```python
allData=allData[['Date','Ticker','Close']]
allData
```

```python
allData.to_csv('MarketData.csv', index=False)
```
